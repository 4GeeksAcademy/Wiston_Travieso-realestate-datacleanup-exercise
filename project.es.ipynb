#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

```python
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Normalizar variable de precio para cada población del cinturón sur de Madrid
ds_cinturón_sur_normalized = ds_cinturón_sur.copy()

for ciudad in ciudades_cinturón_sur:
    ciudad_data = ds_cinturón_sur[ds_cinturón_sur["level5"]==ciudad]
    scaler = StandardScaler()
    ciudad_data["price_normalized"] = scaler.fit_transform(ciudad_data[["price"]])
    ds_cinturón_sur_normalized.loc[ciudad_data.index, "price_normalized"] = ciudad_data["price_normalized"] 
    # Graficar los 4 histogramas de precios normalizados en un mismo gráfico    
    plt.hist(ciudad_data["price_normalized"], bins=30, alpha=0.5, label=ciudad)     
    plt.title("Histogramas de precios normalizados en el cinturón sur de Madrid")
    plt.legend()            
    
```

#### Ejercicio 19. ¿Qué puedes decir sobre el precio por metro cuadrado (precio/m2) entre los municipios de 'Getafe' y 'Alcorcón'? Debes usar el subconjunto obtenido en la pregunta 14 (★★☆)

Pista: Crea una nueva columna llamada `pps` (price per square en inglés) y luego analiza los valores

```python
# Calcular precio por metro cuadrado (pps) para Getafe y Alcorcón
ds_getafe_alacorcón = ds_cinturón_sur[ds_cinturón_sur["level5"].isin(["Getafe", "Alcorcón"])].copy()
ds_getafe_alacorcón["pps"] = ds_getafe_alacorcón["price"]/ds_getafe_alacorcón["surface"]    

# Analizar los valores de precio por metro cuadrado (pps) en Getafe y Alcorcón
precio_promedio_pps_getafe = ds_getafe_alacorcón[ds_getafe_alacorcón["level5"]=="Getafe"]["pps"].mean()
print(f"El precio promedio por metro cuadrado en Getafe es de USD {precio_promedio_pps_getafe}")
precio_promedio_pps_alcorcón = ds_getafe_alacorcón[ds_getafe_alacorcón["level5"]=="Alcorcón"]["pps"].mean()
print(f"El precio promedio por metro cuadrado en Alcorcón es de USD {precio_promedio_pps_alcorcón}")

diferencia_pps = precio_promedio_pps_getafe - precio_promedio_pps_alcorcón
print(f"La diferencia de precio por metro cuadrado entre Getafe y Alcorcón es de USD {diferencia_pps}")
```

#### Ejercicio 21. Realiza un trazado de las coordenadas (columnas latitud y longitud) del cinturón sur de Madrid por color de cada población (debes usar el subconjunto obtenido del Ejercicio 14) (★★★★)

Ejecuta la siguiente celda y luego comienza a codear en la siguiente. Debes implementar un código simple que transforme las columnas de coordenadas en un diccionario de Python (agrega más información si es necesario) y agrégala al mapa.

```python
from ipyleaflet import Map, basemaps

# Mapa centrado en Madrid
map = Map(center = (40.35, -3.8), zoom = 11.5, min_zoom = 1, max_zoom = 20, 
    basemap=basemaps.OpenStreetMap.Mapnik)
map
```

```python
from ipyleaflet import Map, basemaps, Marker, MarkerCluster
import pandas as pd

ds_cinturón_sur_copy = ds_cinturón_sur.copy()

# Convertir las columnas de latitud y longitud a tipo numérico, manejando errores
ds_cinturón_sur_copy['latitude'] = pd.to_numeric(ds_cinturón_sur_copy['latitude'].astype(str).str.replace(',', '.'), errors='coerce')
ds_cinturón_sur_copy['longitude'] = pd.to_numeric(ds_cinturón_sur_copy['longitude'].astype(str).str.replace(',', '.'), errors='coerce')


# Eliminar filas con valores nulos en latitud o longitud después de la conversión
ds_cinturón_sur_clean = ds_cinturón_sur_copy.dropna(subset=['latitude', 'longitude']).copy()

# Crear marcadores para cada propiedad en el cinturón sur
markers = [Marker(location=(row['latitude'], row['longitude'])) for index, row in ds_cinturón_sur_clean.iterrows()]

# Crear un MarkerCluster para agrupar los marcadores
marker_cluster = MarkerCluster(
    markers=markers
)

# Crear el mapa centrado en Madrid
map = Map(center = (40.35, -3.8), zoom = 11.5, min_zoom = 1, max_zoom = 20,
    basemap=basemaps.OpenStreetMap.Mapnik)


# Agregar el MarkerCluster al mapa
map.add_layer(marker_cluster)

# Mostrar el mapa (si no se muestra automáticamente)
map
```
