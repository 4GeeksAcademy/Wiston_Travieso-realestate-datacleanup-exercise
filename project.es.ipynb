#### Ejercicio 09. ¿Son los precios promedios de "Valdemorillo" y "Galapagar" los mismos? (★★☆)

Imprime ambos promedios y escribe una conclusión sobre ellos.

```python
ds_valdemorillo = ds[ds["address"]=="Valdemorillo"]
precio_promedio_valdemorillo = ds_valdemorillo["price"].mean()
print(f"El Precio Promedio de Valdemorillo es de USD {precio_promedio_valdemorillo}")

ds_galapagar = ds[ds["address"]=="Galapagar"]
precio_promedio_galapagar =ds_galapagar["price"].mean()
print(f"El Precio Promedio de Galapagar es de USD {precio_promedio_galapagar}")
print(f"La diferncia entre ambas localidades es de USD{precio_promedio_galapagar-precio_promedio_valdemorillo}")
```

#### Ejercicio 10. ¿Son los promedios de precio por metro cuadrado (precio/m2) de "Valdemorillo" y "Galapagar" los mismos? (★★☆)

Imprime ambos promedios de precio por metro cuadrado y escribe una conclusión sobre ellos.

Pista: Crea una nueva columna llamada `pps` (*price per square* o precio por metro cuadrado) y luego analiza los valores.

```python
ds_valdemorillo = ds[ds["address"]=="Valdemorillo"].copy()
ds_galapagar = ds[ds["address"]=="Galapagar"].copy()

ds_valdemorillo["pps"]= ds_valdemorillo["price"]/ds_valdemorillo["surface"]
ds_galapagar["pps"]= ds_galapagar["price"]/ds_galapagar["surface"]

precio_promedio_valdemorillo_metro = ds_valdemorillo["pps"].mean()
print(f"El Precio Promedio por metro cuadrado de Valdemorillo es de USD {precio_promedio_valdemorillo_metro}")

precio_promedio_galapagar_metro =ds_galapagar["pps"].mean()
print(f"El Precio Promedio por metro de Galapagar es de USD {precio_promedio_galapagar_metro}")

print(f"La diferncia entre ambas localidades es de USD{precio_promedio_galapagar_metro-precio_promedio_valdemorillo_metro}")
```

#### Ejercicio 14. Ahora vamos a trabajar con el "cinturón sur" de Madrid. Haz un subconjunto del DataFrame original que contenga las siguientes poblaciones (columna level5): "Fuenlabrada", "Leganés", "Getafe", "Alcorcón" (★★☆)

Pista: Filtra el DataFrame original usando la columna `level5` y la función `isin`.

```python
ds_cinturón_sur = ds[ds["level5"].isin(["Getafe", "Leganés", "Fuenlabrada", "Alcorcón",])]
print(ds_cinturón_sur.head())
```

#### Ejercicio 15. Traza un gráfico de barras de la mediana de los precios y explica lo que observas (debes usar el subconjunto obtenido del Ejercicio 14) (★★★)

Imprima un gráfico de barras de la mediana de precios y escriba en la celda Markdown un breve análisis sobre el gráfico.

```python
ciudades_cinturón_sur = ds_cinturón_sur["level5"].unique()
print(f"Las ciudades que conforman el cinturón sur de Madrid son: {ciudades_cinturón_sur}")
mediana_precios_cinturón_sur = ds_cinturón_sur.groupby("level5")["price"].median()
print(f"La mediana de precios en el cinturón sur de Madrid es de USD {mediana_precios_cinturón_sur}")

plt.figure(figsize=(10,5))
plt.xlabel("Ciudades")
plt.ylabel("Precios")
mediana_precios_cinturón_sur.plot(kind='bar')
plt.title("Mediana de precios en el cinturón sur de Madrid")
plt.show()
```

En el gráfico podemos observar que en la ciudad de Getafe la media de precios se encuentra considerablemente por encima de los demas, siendo Fuenlabrada y Leganés los más accesibles

#### Ejercicio 16. Calcula la media y la varianza de muestra para las siguientes variables: precio, habitaciones, superficie y baños (debes usar el subconjunto obtenido del Ejercicio 14) (★★★)

Imprime ambos valores por cada variable.
